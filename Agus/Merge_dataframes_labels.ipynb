## Merge dataframes with outliers and cluster labels

```python
import pandas as pd
```

```python
data_with_outliers_labels = pd.read_csv('../data/transactions_with_outliers.csv')

data_with_cluster_labels = pd.read_csv('../data/df_labels_all_methods.csv', index_col=0)
```

```python
data_with_outliers_labels.shape
```

```python
cluster_labels = data_with_cluster_labels[['KMeans_label', 'labels_hierarchical_3c', 'labels_hierarchical_9c']]
```

```python
data_all_labels = pd.concat([data_with_outliers_labels,cluster_labels], axis=1)
data_all_labels.shape
```

```python
data_all_labels.to_csv('../data/labelled_transactions.csv', index=False)
data_all_labels
```
